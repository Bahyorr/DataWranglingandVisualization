### Cleaning Data

1. There are invalid dogs name e.g., "a", "such", "quite", "None", etc.
I resolved this issue by removing all dogs name with lower case and replace all "None" with np.nan in the dataframe.

2. Timestamp column appears to be a string, instead of datetime object.
Here, I converted timestamp column to datetime object and change the data type of id column to string using pandas.to_datetime() function

3. Some ratings are wrongly represented. For instance, instead of 13.5/10, 5/10 was used, instead of 9.75/10, 75/10 was used, instead of 11.27/10, 27/10 was used and instead of 9.5/10, 5/10 in the data frame.
Here, I was able to extract both ratings numerator and denominator from the text column, and also use astype() method to change their type from string to float.

4. There are misrepresentation of NA values in different column such as Doggo, Floofer, Pupper, Puppo etc.
I used dataframe.replace() function to replace all the "None" strings in the dog_stages columns

5. Appearance of unknown url at the end of some tweets. I was able to remove the url at the end of some tweets using replace() method in str function

6. There are invalid huge ratings numerator and denominator. I used pandas query() method to extract dataframe with ratings denominator of 10 and ratings numerator <= 14. I will also evaluate the rating_numerator column by dividing it by 10.

7. Retweeted tweets and replies not needed. I performed the action of removing all retweeted tweets and replies

8. Retweet columns and replies not needed. I droppped all retweeted tweets and replies columns using drop() method in pandas
